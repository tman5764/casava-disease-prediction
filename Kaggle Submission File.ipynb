# Load test data sets, and model

This section of the program is just loading the data sets and the model. 

```python
test = pd.read_csv("/content/drive/MyDrive/Prediction of Disease in Cassava Plants. /Cassava/sample_submission.csv",dtype=str)
test_path = "/content/drive/MyDrive/Prediction of Disease in Cassava Plants. /Cassava/test_images"
print(test.head())
```

```python
new_model = keras.models.load_model('/content/drive/MyDrive/Prediction of Disease in Cassava Plants. /my_model_retrain_v01.h5')
new_model.summary()
```

# Data Generator for the test images. 

```python
test_datagen = ImageDataGenerator(rescale=1/255)

bs = 256

test_generator = test_datagen.flow_from_dataframe(
    dataframe = test,
    directory = test_path,
    x_col = "image_id",
    y_col = None,
    batch_size = bs,
    seed = 1,
    shuffle = False,
    class_mode = None,
    target_size = (150,200))

te_size = 1

te_steps = math.ceil(len(test_generator) / bs)
```

# Make predictions on the testing images

This section of the notebook will be taking the testing images and will be making predictions on them. it will then be compiling the data into a CSV file that will be used to make the predictions. 

```python
test_pred = new_model.predict(test_generator)
```

```python
test_fnames = test_generator.filenames
pred_classes = np.argmax(test_pred, axis=1)
```

```python
submission = pd.DataFrame({
    'image_id':test_fnames,
    'label':pred_classes
})
```

```python
submission.to_csv('submission.csv', index=False)
```
